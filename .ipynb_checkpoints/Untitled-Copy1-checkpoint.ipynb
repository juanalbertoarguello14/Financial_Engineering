```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
%matplotlib inline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
ticker = "SPY"
df = yf.download(ticker, start="2015-01-01", end="2026-01-01")
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)
df['Log_Ret'] = np.log(df[price_col] / df[price_col].shift(1))
df['Vol_21d'] = df['Log_Ret'].rolling(window=21).std() * np.sqrt(252)

df['MA_Ratio'] = df[price_col].rolling(5).mean() / df[price_col].rolling(20).mean()
df['RSI'] = 100 - (100 / (1 + (df['Log_Ret'].where(df['Log_Ret'] > 0, 0).rolling(14).mean() / 
                               df['Log_Ret'].where(df['Log_Ret'] < 0, 0).abs().rolling(14).mean())))
df['Target_Vol'] = df['Vol_21d'].shift(-21)
df.dropna(inplace=True)
```

```python
df['Returns'] = np.log(df['Close'] / df['Close'].shift(1))
df['Volatility'] = df['Returns'].rolling(window=21).std() * np.sqrt(252)
df['Vol_Lag1'] = df['Volatility'].shift(1)
df['Vol_Lag2'] = df['Volatility'].shift(2)
```

```python
plt.figure(figsize=(12, 6))
df['Volatility'].plot()
plt.title('21-Day Rolling Realized Volatility (SPY)')
plt.ylabel('Annualized Volatility')
plt.show()
```

```python
X = df[['Vol_Lag1', 'Vol_Lag2', 'RSI', 'MA_Ratio']].dropna()
y = df['Target_Vol'].loc[X.index]

split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

model = RandomForestRegressor(n_estimators=200, max_depth=7, random_state=42)
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': preds}, index=y_test.index)

comparison.plot(figsize=(12,6), alpha=0.8)
plt.title('Random Forest: Predicted vs Actual Volatility')
plt.show()
```
