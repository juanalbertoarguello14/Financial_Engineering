```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# -----------------------------
# Data Download
# -----------------------------
ticker = "SPY"
df = yf.download(ticker, start="2015-01-01", end="2026-01-01")

if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)

price_col = "Close"

# -----------------------------
# Feature Engineering
# -----------------------------
df["Log_Ret"] = np.log(df[price_col] / df[price_col].shift(1))

df["Vol_21d"] = df["Log_Ret"].rolling(21).std() * np.sqrt(252)

df["Vol_Lag1"] = df["Vol_21d"].shift(1)
df["Vol_Lag2"] = df["Vol_21d"].shift(2)

df["MA_Ratio"] = (
    df[price_col].rolling(5).mean() /
    df[price_col].rolling(20).mean()
)

gain = df["Log_Ret"].clip(lower=0)
loss = -df["Log_Ret"].clip(upper=0)

avg_gain = gain.rolling(14).mean()
avg_loss = loss.rolling(14).mean()

rs = avg_gain / avg_loss
df["RSI"] = 100 - (100 / (1 + rs))

# Target: 21-day forward volatility
df["Target_Vol"] = df["Vol_21d"].shift(-21)

df.dropna(inplace=True)

# -----------------------------
# Train / Test Split 
# -----------------------------
features = ["Vol_Lag1", "Vol_Lag2", "RSI", "MA_Ratio"]

X = df[features]
y = df["Target_Vol"]

split = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]

# -----------------------------
# Model
# -----------------------------
model = RandomForestRegressor(
    n_estimators=300,
    max_depth=7,
    min_samples_leaf=10,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)

# -----------------------------
# Predictions & Evaluation
# -----------------------------
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Model Performance:")
print(f"MSE  : {mse:.6f}")
print(f"RMSE : {rmse:.6f}")
print(f"RÂ²   : {r2:.4f}")

# -----------------------------
# Plot Actual vs Predicted
# -----------------------------
comparison = pd.DataFrame(
    {"Actual": y_test, "Predicted": y_pred},
    index=y_test.index
)

comparison.plot(figsize=(12, 6), alpha=0.8)
plt.title("Random Forest: Predicted vs Actual 21-Day Volatility (SPY)")
plt.ylabel("Annualized Volatility")
plt.show()

# -----------------------------
# Feature Importance
# -----------------------------
importances = pd.Series(
    model.feature_importances_,
    index=features
).sort_values(ascending=False)

importances.plot(kind="bar", figsize=(8, 4))
plt.title("Feature Importance")
plt.show()
```
